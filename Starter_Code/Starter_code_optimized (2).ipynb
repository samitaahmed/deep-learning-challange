```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(['EIN','NAME'],axis=1,inplace=True)
```

```python
# Determine the number of unique values in each column.
unique_value_count = application_df.nunique()
print(unique_value_count)
```

```python
# Look at APPLICATION_TYPE value counts for binning
value_count_application_type = application_df['APPLICATION_TYPE'].value_counts()
print(value_count_application_type)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 528
application_types_to_replace = list(value_count_application_type[value_count_application_type< cutoff_value].index)
#print(application_types_to_replace)
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE']=application_df['APPLICATION_TYPE'].replace(app,"Other")
```

```python

application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
value_count_classification = application_df['CLASSIFICATION'].value_counts()
print(value_count_classification)
```

```python
# Filter the value counts to show only counts greater than 1
classification_counts = application_df['CLASSIFICATION'].value_counts()


classification_counts_gt1 = classification_counts[classification_counts > 1]

print(classification_counts_gt1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value_classification = 1883
classifications_to_replace = list(classification_counts[classification_counts< cutoff_value_classification].index)
#print(application_types_to_replace)
for app in classifications_to_replace:
    application_df['CLASSIFICATION']=application_df['CLASSIFICATION'].replace(app,"Other")
```

```python
 application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
categorical_columns = ['APPLICATION_TYPE', 'CLASSIFICATION','AFFILIATION','USE_CASE','ORGANIZATION','SPECIAL_CONSIDERATIONS','INCOME_AMT']

application_df = pd.get_dummies(application_df,columns=categorical_columns)
```

```python
application_df
```

```python
# Split our preprocessed data into our features and target arrays
target_column = 'IS_SUCCESSFUL'
X = application_df.drop(columns= [target_column])
y = application_df[target_column]
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

```

```python
X_scaler = scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.api._v2.keras import activations

# Keras
from tensorflow import keras

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.


nn = tf.keras.models.Sequential()




# First hidden layer
nn.add(Dense(80,activation='relu', input_dim = X_train_scaled.shape[1]))

# Second hidden layer
nn.add(Dense(30,activation = 'relu'))
# Third hidden layer
nn.add(Dense(10,activation = 'relu'))

# Output layer
nn.add(Dense(1, activation = 'softmax'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
nn.fit(
    X_train,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
